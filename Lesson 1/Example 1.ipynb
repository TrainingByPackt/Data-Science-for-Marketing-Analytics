```python
import numpy as np 

import pandas as pd
```

```python
df = pd.DataFrame() 

df['views'] = np.random.normal(100000, scale=10000, size=100) 

df['likes'] = np.random.normal(400, scale=10, size=100) 
   
df = df.astype(int) 
df
```

```python
df = df[df['views']>1000][df['likes'] >10]
df
```
